
# Odds, Log-Odds, and Odds Ratio

This notebook is a bridge from linear regression to logistic regression. It introduces the concept of **odds**, **log-odds**, and **odds ratio**, which are fundamental in logistic models.



## üß† What Are Odds?

Odds are another way of representing probability. Instead of saying "there is a 20% chance of rain," odds express the chance as a ratio of success to failure.

If the probability (P) of an event is 0.2, then:

**Odds = P / (1 - P) = 0.2 / 0.8 = 0.25**

That means: for every 1 time it rains, there are 4 times it doesn't.


```python

def probability_to_odds(p):
    return p / (1 - p)

probability = 0.2
odds = probability_to_odds(probability)
print(f"Probability: {probability}, Odds: {odds}")
```


## üîÅ Probability vs Odds

- **Probability** tells you how likely something is to happen out of all possibilities (between 0 and 1).
- **Odds** tell you how likely it is to happen *relative to not happening* (range: 0 to ‚àû).

| Metric      | Formula             | Example (P = 0.2) |
|-------------|---------------------|-------------------|
| Probability | P                   | 0.2               |
| Odds        | P / (1 - P)         | 0.25              |


```python

import numpy as np

def odds_to_log_odds(odds):
    return np.log(odds)

log_odds = odds_to_log_odds(odds)
print(f"Log-Odds: {log_odds}")
```

## Plot Probability vs Odds

```python

import matplotlib.pyplot as plt

probabilities = np.linspace(0.01, 0.99, 100)
odds_values = probability_to_odds(probabilities)

plt.plot(probabilities, odds_values)
plt.title("Probability to Odds")
plt.xlabel("Probability")
plt.ylabel("Odds")
plt.grid(True)
plt.show()
```


## üßÆ Why Log(Odds)?

Odds values are skewed:
- 1/6 gives 0.166 (close to 0)
- 6/1 gives 6.0 (can grow infinitely)

To **normalize** the scale and enable regression, we apply the natural logarithm:

**Log-Odds = log(P / (1 - P))**

This gives us values ranging from negative infinity to positive infinity, which is perfect for using in a linear regression setting like **logistic regression**.


```python

log_odds_values = odds_to_log_odds(odds_values)

plt.plot(probabilities, log_odds_values)
plt.title("Probability to Log-Odds")
plt.xlabel("Probability")
plt.ylabel("Log-Odds")
plt.grid(True)
plt.show()
```
