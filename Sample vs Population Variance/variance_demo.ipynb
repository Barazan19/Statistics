```python
# variance_demo.ipynb (exportable Jupyter Notebook)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Create population: 15 children with heights from 170 to 184
population = np.arange(170, 185)
population_mean = np.mean(population)
population_variance = np.var(population)

# Draw 100 random samples of size 3
n_samples = 100
sample_size = 3
vars_div_n = []
vars_div_n1 = []

for _ in range(n_samples):
    sample = np.random.choice(population, sample_size, replace=False)
    sample_mean = np.mean(sample)

    var_n = np.sum((sample - sample_mean)**2) / sample_size
    var_n1 = np.sum((sample - sample_mean)**2) / (sample_size - 1)

    vars_div_n.append(var_n)
    vars_div_n1.append(var_n1)

# Mean of sample variances
mean_var_n = np.mean(vars_div_n)
mean_var_n1 = np.mean(vars_div_n1)

# Display results
print("Population Mean:", population_mean)
print("Population Variance:", population_variance)
print("Mean of Sample Variances (divided by n):", mean_var_n)
print("Mean of Sample Variances (divided by n - 1):", mean_var_n1)

# Plot histograms
plt.figure(figsize=(10,6))
plt.hist(vars_div_n, bins=30, alpha=0.6, label='Sample Variance / n')
plt.hist(vars_div_n1, bins=30, alpha=0.6, label='Sample Variance / (n - 1)')
plt.axvline(population_variance, color='red', linestyle='dashed', linewidth=2, label='True Population Variance')
plt.title("Sample Variance Estimation")
plt.xlabel("Variance")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Save to DataFrame (optional for export)
result_df = pd.DataFrame({
    'Sample_Var_Div_n': vars_div_n,
    'Sample_Var_Div_n-1': vars_div_n1
})

result_df.to_csv("sample_variance_results.csv", index=False)
```
