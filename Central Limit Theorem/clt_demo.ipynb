```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# Uniform distribution
uniform_pop = np.random.uniform(0, 100, 100000)

# Exponential distribution
expon_pop = np.random.exponential(scale=50, size=100000)

def sample_means(population, sample_size=30, num_samples=1000):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        means.append(np.mean(sample))
    return means

# Get sample means
uniform_means = sample_means(uniform_pop)
expon_means = sample_means(expon_pop)

# Plot
fig, axes = plt.subplots(1, 4, figsize=(18, 5))

sns.histplot(uniform_pop, bins=50, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Original Uniform Distribution')

sns.histplot(expon_pop, bins=50, kde=True, ax=axes[1], color='red')
axes[1].set_title('Original Exponential Distribution')

sns.histplot(uniform_means, bins=50, kde=True, ax=axes[2], color='green')
axes[2].set_title('Sample Means from Uniform (CLT)')

sns.histplot(expon_means, bins=50, kde=True, ax=axes[3], color='orange')
axes[3].set_title('Sample Means from Exponential (CLT)')

plt.tight_layout()
plt.show()
```
